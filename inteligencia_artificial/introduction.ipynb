```python
clb = df.pop("Club")

df = pd.concat([df.reset_index(drop=True), pd.get_dummies(clb, prefix='clb').reset_index(drop=True)], axis=1, sort=False)
```

```python
def value_to_float(x):
    """
    From K and M to float.
    
    """
    x = x.replace('€', '')
    ret_val = 0.0
    
    if type(x) == float or type(x) == int:
        ret_val = x
    if 'K' in x:
        if len(x) > 1:
            ret_val = float(x.replace('K', ''))
        ret_val = ret_val *1000
    if 'M' in x:
        if len(x) > 1:
            ret_val = float(x.replace('M', ''))
        ret_val = ret_val * 1000000.0
    return ret_val
```

```python
df["Value"] = df["Value"].apply(value_to_float)
df["Wage"] = df["Wage"].apply(value_to_float)
```

```python
def nationality_encoding(x):
    
    sol = 0
    if 'Spain' in x:
        sol = 1
    if 'France' in x:
        sol = 1
    if 'Brazil' in x:
        sol = 1
    if 'England' in x:
        sol = 1
    if 'Argentina' in x:
        sol = 1
    if 'Germany' in x:
        sol = 1
    if 'Italy' in x:
        sol = 1
    if 'Netherlands' in x:
        sol = 1
    if 'Portugal' in x:
        sol = 1
    if 'Belgium' in x:
        sol = 1
    if 'Colombia' in x:
        sol = 1
    return sol
```

```python
df["Nationality"] = df["Nationality"].apply(nationality_encoding)
```

```python
def height_to_cm(x):
  
    x = x.replace("'", '.')
    sol = float(x)
    sol = sol * 30.48
    return sol
```

```python
df["Height"] = df["Height"].apply(height_to_cm)
```

```python
val = df.pop("Value")
to_drop = [c for c in df.columns if c.startswith('Name') or c.startswith('Preferred') or c.startswith('Position')]

df = df.drop(to_drop, axis=1)
```

```python
X_train, X_test, y_train, y_test = train_test_split(df, val, test_size=0.33, random_state=42)
```
